```python
import pandas as pd
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df=pd.read_csv(r"C:\Users\SUHANI\Downloads\cancerdataset1.csv")
df['Patient Id']=le.fit_transform(df['Patient Id'])
df['Age']=le.fit_transform(df['Age'])
df['Gender']=le.fit_transform(df['Gender'])
df['Air Pollution']=le.fit_transform(df['Air Pollution'])
df['Alcohol use']=le.fit_transform(df['Alcohol use'])
df['Dust Allergy']=le.fit_transform(df['Dust Allergy'])
df['OccuPational Hazards']=le.fit_transform(df['OccuPational Hazards'])
df['Genetic Risk']=le.fit_transform(df['Genetic Risk'])
df['chronic Lung Disease']=le.fit_transform(df['chronic Lung Disease'])
df['Balanced Diet']=le.fit_transform(df['Balanced Diet'])
df['Fatigue']=le.fit_transform(df['Fatigue'])
df['Weight Loss']=le.fit_transform(df['Weight Loss'])
df['Shortness of Breath']=le.fit_transform(df['Shortness of Breath'])
df['Wheezing']=le.fit_transform(df['Wheezing'])
df['Swallowing Difficulty']=le.fit_transform(df['Swallowing Difficulty'])
df['Clubbing of Finger Nails']=le.fit_transform(df['Clubbing of Finger Nails'])
df['Frequent Cold']=le.fit_transform(df['Frequent Cold'])
df['Dry Cough']=le.fit_transform(df['Dry Cough'])
df['Snoring']=le.fit_transform(df['Snoring'])
df['int-level']=le.fit_transform(df['int-level'])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='int-level', data=df,)
plt.title('Target Distribution')
```

```python
def plot(col, df=df):
    return df.groupby(col)['int-level'].value_counts(normalize=True).unstack().plot(kind='bar', figsize=(8,5))
plot('Gender')
```

```python
df_new=df.drop(columns=['Gender','Age', 'Air Pollution', 'Shortness of Breath'])
df_new
```

```python
#Finding Correlation
cn=df_new.corr()
cn
```

```python
#Correlation 
cmap=sns.diverging_palette(260,-10,s=50, l=75, n=6,as_cmap=True)
plt.subplots(figsize=(18,18))
sns.heatmap(cn,cmap=cmap,annot=True, square=True)
plt.show()
```

```python
df_new['ALCDUSALL']=df_new['Alcohol use']*df_new['Dust Allergy']
df_new
```

```python
#Splitting independent and dependent variables
X = df_new.drop('int-level', axis = 1)
y = df_new['int-level']
```

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)
```

```python
#Fitting training data to the model
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
#Model accuracy
from sklearn.metrics import classification_report, accuracy_score, f1_score
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

```python
#Decision tree
#Fitting training data to the model
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)  
dt_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

```python
#Fitting K-NN classifier to the training set  
from sklearn.neighbors import KNeighborsClassifier  
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knn_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_knn_pred= knn_model.predict(X_test)
y_knn_pred
```

```python
#Model accuracy
knn_cr=classification_report(y_test, y_knn_pred)
print(knn_cr)
```

```python
#Fitting Gaussian Naive Bayes classifier to the training set  
from sklearn.naive_bayes import GaussianNB
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_gnb_pred= gnb_model.predict(X_test)
y_gnb_pred
```

```python
#Model accuracy
gnb_cr=classification_report(y_test, y_gnb_pred)
print(gnb_cr)
```
